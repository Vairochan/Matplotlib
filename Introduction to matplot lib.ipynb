### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10, 5))
#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randn(100), np.random.randn(100));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10, 5))
#Plot to each different index
ax[0, 0].plot(x, x/2); 
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(100));
```

# OO Method

### Which one should you use?( pyplot vs matplotlib OO method?)

* When ploting something quickly, okay to use pyplot method
* When ploting something more advance use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot Method
over_50.plot(figsize=(20, 8),kind='scatter',
            x='age',
             y='chol',
             c='target');
```

```python
# OO Method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
                 x='age',
                 y='chol',
                 c='target',
                 ax=ax);
```

```python
## OO Method from strach
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
# Customize the plot
ax.set(title="Herat Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholestrol")
#Add a Legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

#add data to ax0 
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
# Customize ax0
ax0.set(title="Heart Disease and cholestrol levels",
       ylabel="Cholestrol");
#ADD a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# ADD a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# add label
ax1.set(title="Heart Disease and Max heart Rate",
       xlabel="Age",
       ylabel="Max heart rate");
# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean Line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to a figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Customising the y and x limitations

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

#add data to ax0 
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap="winter")
# Customize ax0
ax0.set(title="Heart Disease and cholestrol levels",
       ylabel="Cholestrol");

#set x axis limits
ax0.set_xlim([50, 80])

#ADD a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# ADD a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
# add label
ax1.set(title="Heart Disease and Max heart Rate",
       xlabel="Age",
       ylabel="Max heart rate");

ax1.set_ylim([60, 200])
# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean Line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to a figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
